```python
import numpy as np
from keras.layers import LSTM, Dense
from keras.models import Sequential

# Define the training data
x_train = np.array([[1, 2, 3], [2, 3, 4], [3, 4, 5], [4, 5, 6]])
y_train = np.array([4, 5, 6, 7])

# Define the model
model = Sequential()
model.add(LSTM(units=32, input_shape=(3, 1)))
model.add(Dense(units=1))
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['accuracy'])

# Reshape the data for the LSTM layer
x_train = x_train.reshape((x_train.shape[0], x_train.shape[1], 1))

# Train the model
model.fit(x_train, y_train, epochs=100, verbose=1)

# Predict the location of the ball
x_test = np.array([[5, 6, 7]])
x_test = x_test.reshape((x_test.shape[0], x_test.shape[1], 1))
prediction = model.predict(x_test)
print("Predicted location of the ball: ", prediction)
```
